# **Learn Pandas basic using simple data**


```python

# Create a dictionary representing software developers which have 10 element which represent 10 columns and
#  have their ow data which represent the row
#  at the max of them is 5 that mean we can create data frame which have 5 rows and 10 columns

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 28, 35, 24],
    "Experience (Years)": [3, 7, 5, 10, 2],
    "Primary Language": ["Python", "JavaScript", None, "Go", "Python"],
    "Secondary Language": ["SQL", "TypeScript", "C++", "Python", "HTML"],
    "Current Role": ["Backend", "Frontend", "Full Stack", "DevOps", "Data Analyst"],
    "Salary ($)": [85000, 95000, 92000, 105000, 78000],
    "Location": ["NY", "SF", "LA", "Chicago", "Boston"],
    "Remote Work": [True, True, False, False, True],
    "Certifications": [2, 1, 3, 4, None]
}
#  create data frame from the above dict

df = pd.DataFrame(data)
df.info()
df.describe()

```

```python
#  accessing columns
#  we can use df.columns it return index of all column name as Index type
#  we can covert to list and iterate over them  and treat them as python list

list(df.columns)
```

```python
# df.values it return numpy array as the values which hve dimension like row x column
table = df.values
table.shape
#  we can access the single column using slicing as we do in normal np array


```
